```python
import requests
import json

url = 'https://api.lens.org/patent/search'
data = '''{
    "query": {
        "bool": {
            "must": [
                {
                    "range": {
                        "created": {
                            "gt": "2024-11-11"
                        }
                    }
                }
            ]
        }
    }
}'''
headers = {'Authorization': '', 'Content-Type': 'application/json'}
response = requests.post(url, data=data, headers=headers)
if response.status_code != requests.codes.ok:
    print(f"Error: {response.status_code}")
else:
    try:
        # Attempt to pretty-print JSON response
        response_json = response.json()  # Parse response text as JSON
        pretty_text = json.dumps(response_json, indent=4, ensure_ascii=False)  # Pretty-print JSON
        print(pretty_text)
    except ValueError:
        # If response is not JSON, print raw text
        print("Response is not in JSON format. Raw response:")
        print(response.text)
```
